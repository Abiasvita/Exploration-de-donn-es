```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
covid= pd.read_csv('covid_19.csv')
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='New deaths',data=covid)
plt.title('Repartition de Nouveaux dcs')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.histplot(data=covid,x='Recovered / 100 Cases',hue='WHO Region',kde=True,bins=30)
plt.title('guerison selon le region OMS')
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.boxplot(y='1 week % increase',hue='WHO Region',data= covid)
plt.title('region OMS par Pourcentage daugmentation sur 1 semaine')
plt.show()
```
